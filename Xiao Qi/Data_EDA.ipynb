```python
import pandas as pd
import matplotlib
import plotly
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('terrorism_new.csv')
# df = df.loc[:,['motive', 'crit1', 'crit2', 'crit3', 'attacktype1', 'attacktype1_txt', 'attacktype2', 'attacktype2_txt', 'attacktype3', 'attacktype3_txt', 'targtype1', 'targtype1_txt', 'targsubtype1', 'targsubtype1_txt', ]]
df['attacktype1_txt'] = df['attacktype1_txt'].astype('category')
df['targtype1_txt'] = df['targtype1_txt'].astype('category')
df['casualitiestotal']=df['nkill']+df['nwound']
df['nkillvic'] = df['nkill'] - df['nkillter']
df['nwoundvic'] = df['nwound'] - df['nwoundte']
df.head()
```

```python
print(df.loc[df['INT_LOG'] != -9]['INT_LOG'].count() / len(df))
print(df.loc[df['INT_IDEO'] != -9]['INT_IDEO'].count() / len(df))
print(df.loc[df['INT_MISC'] != -9]['INT_MISC'].count() / len(df))
```

```python
coun_terror=df['region_txt'].value_counts().to_frame()
coun_terror.columns=['Attacks']
coun_kill=df.groupby('region_txt')['nkillvic'].sum().to_frame()
coun_wound = df.groupby('region_txt')['nwoundvic'].sum().to_frame()
count_terror_kill = coun_terror.merge(coun_kill,left_index=True,right_index=True,how='left')
count_terror_kill.merge(coun_wound,left_index=True,right_index=True,how='left').plot.bar(width=0.9)

fig=plt.gcf()
fig.set_size_inches(18,6)
plt.title("Victim casualties across Regions")
plt.show()
```
