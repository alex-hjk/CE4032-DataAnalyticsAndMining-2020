```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
import seaborn as sns
```

```python
def removeUnknown(df_in, col, match):
    df = df_in.copy()
    if (match == "Unknown"):
        df = df.loc[df[col].astype(str) != match]
    else: 
        df = df.loc[df[col] != match]
    return df

df = pd.read_csv('terrorism_new.csv')
# cols = ['iyear', 'extended', 'latitude', 'longitude', 'attacktype1', 'success', 'suicide', 'targtype1', 'gname', 'nperps', 'nkill', 'nwound', 'property']
cols = ['iyear', 'extended', 'country', 'attacktype1', 'success', 'suicide', 'targtype1', 'gname', 'nperps', 'nkill', 'nwound', 'property']

df = df.loc[:, cols]

#remove all unknown
for col in cols:
    df = removeUnknown(df, col, 'Unknown')
    df = removeUnknown(df, col, -9)
    df = removeUnknown(df, col, -99)
    
df = df.dropna()
df
# df.to_csv('trimmed_dataset.csv')
```

```python
from sklearn.datasets import make_friedman1
from sklearn.feature_selection import RFE
from sklearn.preprocessing import LabelEncoder

from sklearn.svm import SVR
import numpy as np
from sklearn.model_selection import train_test_split

from sklearn.naive_bayes import CategoricalNB

gnameNum = LabelEncoder().fit_transform(df['gname'])
df['gname'] = gnameNum

classes = df['targtype1']
attributes = df.drop(['targtype1'],axis=1)

X_train, X_test, y_train, y_test = train_test_split(attributes, classes, test_size=0.3, random_state=8)

clf = CategoricalNB()
# clf.fit(X_train, y_train)
# clf.coef_.ndim

from sklearn.ensemble import RandomForestClassifier
# Instantiate model with 1000 decision trees
forest_classifier = RandomForestClassifier(n_estimators = 100)
# Train the model on training data
forest_classifier.fit(X_train, y_train);


attr = getattr(forest_classifier, 'feature_importances_', None)
type(attr)

# selector = RFE(clf, n_features_to_select=5, step=1)
# selector.fit(X_train, y_train)
# selector.support_
```
